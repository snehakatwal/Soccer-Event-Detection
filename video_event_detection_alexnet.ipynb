```python
from keras.models import load_model
import numpy as np
import cv2
```

```python
model = load_model("/content/drive/MyDrive/Major Project/Model/model_AlexNet_V1.h5")

writer = None
(Width, Height) = (None, None)

output_video = '/content/drive/MyDrive/Major Project/Output Video/demo_output_alexnet_V1.mp4'
```

```python
import tensorflow as tf
from google.colab.patches import cv2_imshow

capture_video = cv2.VideoCapture('/content/drive/MyDrive/Major Project/Copy of 2 combined.mp4')

writer = None
(Width, Height) = (None, None)

frameRate = capture_video.get(cv2.CAP_PROP_FPS)

classes = ['Card', 'Goal', 'Substitution']

while True:
  ret, frame = capture_video.read()
  if not ret:
    break
  if Width is None or Height is None:
    (Width, Height) = (int(capture_video.get(cv2.CAP_PROP_FRAME_WIDTH)),
                       int(capture_video.get(cv2.CAP_PROP_FRAME_HEIGHT)))
  output = frame.copy()
  frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
  frame = cv2.resize(frame, (227,227)).astype('float32')
  frame = frame/255
  frame = np.array(frame).reshape(227,227,3)
  preds = model.predict(np.expand_dims(frame, axis=0))
  confidence = np.max(preds)
  print(confidence)
  if confidence > 0.9:
    i = np.argmax(preds)
    label = classes[i]
    print(label)
  else:
    label = 'None'
    print(label)

  if writer is None:
    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    writer = cv2.VideoWriter(output_video, fourcc,frameRate, (Width, Height), True)

  if label != 'None':
    writer.write(output)

  key = cv2.waitKey(1) & 0xFF

  if key == ord('q'):
    break

print("Finalizing.....")
writer.release()
capture_video.release()
```
