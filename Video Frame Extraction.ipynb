```python
import cv2
import os
import numpy as np
import math

classes = ['None']

set_path = '/content/drive/MyDrive/Major Project/Dataset'
processed_path = '/content/drive/MyDrive/Major Project'

def video_capturing_function(dataset,folder_name):

    count = 0
    i = 0
    # for direc in os.listdir(dataset):
    #   dataset_path = os.path.join(dataset,direc)

    for video in os.listdir(dataset):
        video_read_path=os.path.join(dataset,video)
        cap=cv2.VideoCapture(video_read_path)
        cap.set(cv2.CAP_PROP_FPS, 1)

        frameRate = cap.get(5)

        if not os.path.exists(os.path.join(os.path.join(processed_path,folder_name),classes[i])):
          os.mkdir(os.path.join(os.path.join(processed_path,folder_name),classes[i]))

        train_write_file = os.path.join(os.path.join(processed_path,folder_name),classes[i])

        if not cap.isOpened():
          print("Error opening video file")
          exit()

        while(cap.isOpened()):
            frameId = cap.get(1)
            ret, frame = cap.read()

            if (ret != True):
                break

            if (frameId % math.floor(frameRate/15) == 0):
                filename = f"{count}.jpg"
                count += 1
                # frame_RGB = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                cv2.imwrite(os.path.join(train_write_file , filename), frame)

    i += 1

    cap.release()
    cv2.destroyAllWindows()

    return print(f"All frames written in the: {folder_name} Folder")

video_capturing_function(set_path,"Preprocessed Dataset")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
import cv2
import os
import numpy as np
import math
from keras.preprocessing.image import ImageDataGenerator
from keras.models import load_model

model = load_model("/content/drive/MyDrive/Major Project/Model/model_AlexNet_V1.h5")


# Create an ImageDataGenerator object
datagen = ImageDataGenerator(
    validation_split = 0.2,
    rotation_range=40,  # Degrees of random rotation
    width_shift_range = 0.2,  # % shift in horizontal direction
    height_shift_range = 0.2,  # % shift in vertical direction
    shear_range = 0.2,  # Shear angle in counter-clockwise direction in degrees
    rescale = 1./255, # Standardization
    horizontal_flip=True,  # Horizontal flipping
    fill_mode='nearest'  # Fill missing pixels using the 'nearest' pixel
)


# Define the input directory and batch size
input_dir_test = '/content/drive/MyDrive/Major Project/Test'

# Generate augmented image data
image_generator_test = datagen.flow_from_directory(
    input_dir_test,
    batch_size=32,
    subset = 'training',
    target_size=(227, 227),  # Desired image size
    class_mode='categorical'  # Set the class mode based on your problem
    )
```

```python
evaluation = model.evaluate(image_generator_test)
print(evaluation)
```

```python
# predict probabilities for test set
yhat_probs = model.predict(image_generator_test, verbose=0)
# predict crisp classes for test set
yhat_classes = np.argmax(yhat_probs, axis=1)
```
